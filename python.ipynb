#perplexity code :: 
# NOTE:
# 1) bernoulli_numbers(n_max): recursion से B_0 ... B_{n_max} निकालता है.
# 2) tan_series_bernoulli(x, n_terms):
#    Bernoulli numbers और Taylor series से tan(x) को n_terms तक approx करता है.

import math

def bernoulli_numbers(n_max):
    """
    B_0 से B_{n_max} तक Bernoulli numbers निकालता है
    recurrence:
    B_0 = 1
    B_m = -1/(m+1) * sum_{k=0}^{m-1} C(m+1, k) * B_k
    """
    B = [0.0] * (n_max + 1)
    B[0] = 1.0

    for m in range(1, n_max + 1):
        s = 0.0
        for k in range(0, m):
            s += math.comb(m + 1, k) * B[k]
        B[m] = -s / (m + 1)
    return B  # list: B[0], B[1], ..., B[n_max]


def tan_series_bernoulli(x, n_terms):
    """
    tan(x) ≈ sum_{n=1}^{n_terms} [(-1)^{n-1} * 2^{2n} * (2^{2n}-1) * B_{2n} / (2n)!] * x^{2n-1}
    x radians में होना चाहिए
    n_terms = कितने series terms (n) लेने हैं
    """
    # हमें B_{2n} तक चाहिए, यानी B_{2*n_terms}
    B = bernoulli_numbers(2 * n_terms)

    s = 0.0
    for n in range(1, n_terms + 1):
        B_2n = B[2 * n]          # B_{2n}
        coef_num = ((-1)**(n - 1)) * (2**(2 * n)) * (2**(2 * n) - 1) * B_2n
        coef_den = math.factorial(2 * n)
        coef = coef_num / coef_den
        s += coef * (x ** (2 * n - 1))
    return s


if __name__ == "__main__":
    x = float(input("x (radians) दर्ज करो: "))
    n = int(input("कितने terms (n) लेना है?: "))

    approx = tan_series_bernoulli(x, n)
    actual = math.tan(x)

    print("Series approx tan(x) =", approx)
    print("math.tan(x)         =", actual)
    print("Error               =", abs(actual - approx))
